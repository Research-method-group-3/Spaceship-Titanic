```python
# filling missing numerical data with mean/median
# Calculate the skewness of each numeric column
skewness = data.select_dtypes(include=np.number).apply(lambda x: x.skew())

# Fill missing numerical data with mean/median
for col, skew in skewness.items():
    if skew > 1.0 or skew < -1.0:
        fill_value = data[col].median()
        fill_method = 'median'
        print(f"Using median for {col}")
    else:
        fill_value = data[col].mean()
        fill_method = 'mean'
        print(f"Using mean for {col}")
    
    # Fill missing values in the specific column
    data[col].fillna(fill_value, inplace=True)

# Print the updated data
print(data)
```

```python
import re

def groupingCabines():
    cabin_values = []
    for i in categorical_data['Cabin']:
        if isinstance(i, str):  # Check if value is a string
            k1 = 0
            k2 = len(i)
            while k1 < k2:
                if re.search(r'^[a-zA-Z]', i[k1]):
                    cabin_values.append(i[k1].upper())
                    break
                else:
                    k1 += 1
        else:
            cabin_values.append('U')
    return cabin_values

cabin_values = groupingCabines()
cabin_df = pd.DataFrame({'Cabin': cabin_values})
categorical_data['Cabin'] = cabin_df['Cabin']
categorical_data['Cabin'].unique()
```

```python
categorical_data['Destination'].unique()
```

```python
#to fill missing values
# Calculate the percentages of unique values for each categorical column
percentages = categorical_data.apply(lambda col: col.value_counts() / col.value_counts().sum())

# Identify missing values in the categorical columns
missing_mask = categorical_data.isnull()

# Fill missing values based on the percentages of unique values
for featureName in categorical_data.columns:
    missing_count = missing_mask[featureName].sum()
    unique_values = categorical_data[featureName].value_counts().index
    fill_values = np.random.choice(unique_values, size=missing_count, replace=True, p=percentages[featureName].reindex(unique_values, fill_value=0))
    categorical_data.loc[missing_mask[featureName], featureName] = fill_values

# Verify the updated categorical_data DataFrame
print(categorical_data)
print(categorical_data.isnull().sum())
```

```python
categorical_data = pd.get_dummies(categorical_data)

# Print the encoded data
print(categorical_data)
```

```python
# Concatenate the data frames along the columns (axis=1)
X = pd.concat([data, categorical_data], axis=1)
```

```python
X.isnull().sum()
```

```python
correlation_matrix = X.corr()
# Set a threshold for correlation values
threshold = 0.7

# Exclude correlation of value with itself
correlation_matrix = correlation_matrix.mask(np.triu(np.ones(correlation_matrix.shape)).astype(bool))

# Find the highly correlated values
highly_correlated = correlation_matrix.abs().stack().sort_values(ascending=False)
highly_correlated = highly_correlated[highly_correlated > threshold]

# Display the highly correlated values
print(highly_correlated)
```

```python
#Now our data is ready & clean for predictions
from sklearn.model_selection import train_test_split
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0)
```

```python
#Decision Tree
#define DecisionTree model
my_tree= DecisionTreeRegressor()
# Define the SVM model
my_svm = SVC(random_state=1)
#Random Forest Classifier
myRandomForest = RandomForestRegressor(random_state=1)
#Gradient Boosting Classifier:
my_Gradientmodel = GradientBoostingClassifier(random_state=1)

#fit model
my_tree.fit(train_X,train_y)
my_svm.fit(train_X, train_y)
myRandomForest.fit(train_X,train_y)
my_Gradientmodel.fit(train_X, train_y)

#get prediction for validation data
val_prediction1= my_tree.predict(val_X)
val_prediction2= my_svm.predict(val_X)
val_prediction3= myRandomForest.predict(val_X)
val_prediction4= my_Gradientmodel.predict(val_X)
```

```python
val_prediction1 = [round(value) for value in val_prediction1]
#val_prediction1 = [bool(val) for val in val_prediction1]
print(val_prediction1)

# Calculate accuracy
accuracy = accuracy_score(val_y, val_prediction1)
print("Accuracy:", accuracy)
```
